## Bar 

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company, revenue)
```

```python
plt.ylabel('Revenue(Blns)')
plt.title('Top Companies Revenue')
plt.bar(company, revenue)
```

```python
profit=[40,10,14,6]
plt.bar(company, revenue)
plt.bar(company, profit)
```

```python
profit=[40,10,14,6]
plt.bar(company, revenue)
plt.bar(company, profit)
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

# histogram

```python
blood_sugar=[102, 146,  92, 109, 146,  93, 119, 127, 112, 102, 104, 121, 131]
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.99)
```

```python
blood_sugar_men=[102, 146,  92, 109, 146,  93, 119, 127, 112, 102, 104, 121, 131]
blood_sugar_women=[83, 140,  97, 135, 145,  99, 110, 106, 116, 102,  98, 119, 112]
```

```python
plt.title('Blood Sugar Analysis')
plt.xlabel('Sugar level')
plt.ylabel('No. of Patients')
plt.hist([blood_sugar_men, blood_sugar_women],bins=[80,100,125,150], rwidth=0.95, color=['green','orange'], label=['Men','Women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men, blood_sugar_women],bins=[80,100,125,150], rwidth=0.65, color=['green','orange'], label=['Men','Women'],orientation='horizontal')
```

# PIE

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, autopct='%0.1f%%',radius=1.7, shadow=True)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, autopct='%0.1f%%', shadow=True, explode=[0,0,0.1,0.1,0.1])


plt.savefig('piechart.png')
plt.show()
```
