```python
newdf=pd.read_csv('weather_data.csv')
newdf.set_index('day',inplace=True)
newdf
```

```python
ndf=newdf.fillna(method='bfill')
ndf
```

```python
newdf.dropna()
```

```python
nd=newdf.dropna(how="all")
nd
```

```python
avg_temp=nd['temperature'].mean(axis=0)
avg_temp
```

```python
avg_wind=nd['windspeed'].mean(axis=0)
avg_wind
```

```python
nd['temperature'].replace(np.nan, avg_temp, inplace=True)
nd
```

```python
nd['windspeed'].replace(np.nan,avg_wind, inplace=True)
nd
```

```python
nd['event'].value_counts()
```

```python
nd['event'].value_counts().idxmax()
```
