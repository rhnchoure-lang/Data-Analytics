# Tip Calulator
Write a Python program that calculates how much each person should pay when a restaurant bill is split among multiple people, including a tip.

The program should:
1. Display a welcome message.
2. Ask the user for the total bill amount.
3. Ask the user for the tip percentage (10%, 12%, or 15%).
4. Ask how many people will split the bill.
5. Calculate the total bill including the tip.
6. Calculate and display how much each person should pay, rounded to two decimal places.

```python
print("Welcome to Tip & Split Calculator ")
bill=float(input("Enter the total bill amount:- "))
per=float(input("Enter the percentage of tip you would like to give(10%, 12% or 15% ):- "))
people=int(input("How many people will split the bill? "))

tip=(per*bill)/100
tbill=bill + tip
split=tbill / people

print(f"No. of People:- {people}")
print(f"Your Total Bill is(Including tip):-{tbill: .2f}")
print(f'Each Person Should Pay:- {split:.2f}')
print("Thank YOU Visit Again!!")
```
