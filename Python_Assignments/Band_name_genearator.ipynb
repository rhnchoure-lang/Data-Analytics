# Band Name generator

Write a Python program that generates a band name by taking user input.
The program should:  
	1.	Display a welcome message.  
	2.	Ask the user for the name of the city they grew up in.  
	3.	Ask the user for their favourite animal   
	4.	Capitalize both inputs.  
	5.	Display a possible band name by combining the city name and the animal name  

```python
print("Welcome to the BAND NAME Generator.")
city=input("What is the name of the City you grew up in ?").capitalize()
animal=input("What is your Favourite Animal ?")
animal= animal.capitalize()
band=city + " "+animal
print(f"Your Band Name could be {band}!!!")
```
