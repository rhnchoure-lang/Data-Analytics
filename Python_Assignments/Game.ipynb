Pgm3: Write a Python program that simulates a simple text-based adventure game called “Treasure Island.”  
  
The program should:  
	1.	Display a welcome message and explain the mission.  
	2.	Ask the user to choose a direction at a crossroad (left or right).  
	3.	Based on the user’s choice, present further decisions using conditional statements.   
	4.	Convert all user input to lowercase to handle case-insensitive input.  
	5.	End the game with either “You Win!” or “Game Over.” depending on the choices made.  



Example:   
Welcome to Treasure Island.  
Your mission is to find the treasure. 
You're at a cross road. Where do you want to go? Type 'left' or 'right'  
left  
You've come to a lake. There is an island in the middle of the lake. Type 'wait' to wait for a boat. Type 'swim' to swim across.  
wait  
You arrive at the island unharmed. There is a house with 3 doors. One red, one yellow and one blue. Which colour do you choose?  
yellow  
You found the treasure! You Win!  
  

Ex 2:-
You're at a cross road. Where do you want to go? Type 'left' or 'right'
right
You fell into a hole. Game Over.

```python
print("Welcome to the treasure Island")
print("Your mission is to find the treasure. You're at a cross road. Where do you want to go?")
inp1=input("Type'left'or'right'").lower()
if (inp1 == 'left'):
    print("You've come to a lake. There is an island in the middle of the lake. ")
    inp2=input("Type 'wait' to wait for a boat. Type 'swim' to swim across.").lower()
    if (inp2 == 'wait'):
        print("You arrive at the island unharmed. There is a house with 3 doors. One red, one yellow and one blue.")
        inp3=input("Which colour do you choose?").lower()
        if (inp3 == 'yellow'):
            print("You found the treasure! You Win!")
        elif (inp3 == 'blue'):
            print("You died by the bite of the snake came out of the door!! GAME OVER!!")
        else :
            print("There is nothing here!! You lost!!")
    else:
        print("You died while swimming towards island!! GAME OVER!!")
else:
    print("You fell into a hole. Game Over.!!")
```
