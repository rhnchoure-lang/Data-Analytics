## Write a Python program that implements a text-based Hangman game.

The program should:  
	1.	Display a Hangman ASCII-art title at the start.  
	2.	Randomly select a word from a predefined list of famous characters.  
	3.	Display blanks (_) representing each character in the selected word.  
	4.	Allow the user to guess one letter at a time.  
	5.	Reveal the correct letters in their respective positions if the guess is correct.  
	6.	Reduce the number of remaining attempts for each incorrect guess.  
	7.	End the game when the user either guesses t  

```python
import random
word=['rabbit','animal','guess','impressive','lengthy']
choosew=random.choice(word)
display=[]
for i in range(len(choosew)):
    display.append('_')


print('Welcome to the HANGMAN GAME!!')
print(" ".join(display))

chance=8
gameover=False
while not gameover and '_' in display:
    guess=input("Guess the letter :- ").lower()
    if guess in choosew:
        print('Right Guess')
        for j in range(len(choosew)):
            if choosew[j] == guess:
                display[j] = guess
    else:
        chance -= 1
        print("Wrong Guess!!")
        print(f'You have {chance} more guesses left.')

        if chance == 0:
            gameover = True
            print("You LOST!!")
            print("Word was :- ",choosew)

    print(" ".join(display))
    
```
