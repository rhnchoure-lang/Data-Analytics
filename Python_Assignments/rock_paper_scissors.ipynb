# Write a Python program that simulates the Rock, Paper, Scissors game between a user and the computer.

The program should:
	1.	Display instructions for the user to choose Rock, Paper, or Scissors using numbers (0, 1, 2).  
	2.	Take the user’s input and display the corresponding ASCII art.  
	3.	Randomly generate the computer’s choice using the random module.  
	4.	Display the computer’s choice with ASCII art.  
	5.	Determine the winner based on the rules of the game and display the result. 

```python
import random
```

```python
print("Lets Play Rock-Paper-Scissors")
us=int(input("Enter Your choice with 0-Rock, 1-Paper, 2-Scissors"))
comp=random.randint(0,2)



rock="""Rock
        __---------------.
               __________)
              ( ________)
              (________)
              (_______)
        --____(_____)"""
paper="""Paper
         __--------------------.
                _______________)
                ____________)
               ____________)
               ___________)
         --_____________)"""
scissors="""Scissors
               _       ,/'
              (_).  ,/'
               _  ::
              (_)'  `\.
                       `\."""
if(us==0):
    print('Your choice',rock)
    if(comp==0):
        print("Computer choice",rock)
        print("There's a TIE!!!")
    elif(comp==1):
        print("Computer choice:",paper)
        print("You WIN !!!")
    elif(comp==2):
        print("Computer choice:",scissors)
        print("You LOST !!!")
elif(us==1):
    print('Your choice',paper)
    if(comp==0):
        print("Computer choice",rock)
        print("You LOST!!!")
    elif(comp==1):
        print("Computer choice:",paper)
        print("There's a TIE!!!")
    elif(comp==2):
        print("Computer choice:",scissors)
        print("You WIN !!!")
elif(us==2):
    print('Your choice',scissors)
    if(comp==0):
        print("Computer choice",rock)
        print("You LOST !!!")
    elif(comp==1):
        print("Computer choice:",paper)
        print("You WIN !!!")
    elif(comp==2):
        print("Computer choice:",scissors)
        print("There's a TIE!!!")
```
