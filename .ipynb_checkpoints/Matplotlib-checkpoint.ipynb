```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title('Zoomed In');
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([1, 1, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title('Zoomed In');
```

```python
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title('Power of 4')


# Small Axes
axes2 = fig.add_axes([1.1, 0, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title('Zoomed In');
```

```python
# dpi is used for clarity. higher the dpi value, more is the clarity
fig = plt.figure(figsize=(4, 4), dpi=200)

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a, b)
fig.savefig('new_figure.png', bbox_inches = 'tight')
# bbox_inches is used for including the x and y axes
```

### Subplots Functionality

- The plt.subplots() call allows us to easily create figure and axes objects in side by side formations.
- The plt.subplots() command returns a tuple containing the Figure Canvas and then a numpy array holding the axes objects.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0, 10)
y = x*2
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(b, a)
axes[1][1].plot(x, -y)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(x, y)
axes[1].plot(a, b)
axes[2].plot(b, a)

plt.tight_layout()

# using tight layout to give spaces between the axes
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(b, a)
axes[1][1].plot(x, -y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize = (8, 8))

axes[0][0].plot(x, y)
axes[0][1].plot(a, b)
axes[1][0].plot(b, a)
axes[1][1].plot(x, -y)

plt.subplots_adjust(wspace=1, hspace=1)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[0][0].set_title('TITLE 0,0')

axes[0][1].plot(a, b)
axes[0][1].set_title('TITLE 0,1')

axes[1][0].plot(b, a)
axes[1][0].set_title('TITLE 1,0')

axes[1][1].plot(x, -y)
axes[1][1].set_title('TITLE 1,1')

plt.suptitle("Figure Level Title", fontsize = 16)

plt.tight_layout()
```

### Legends

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x, label='X vs X')
axes.plot(x, x**2, label = 'X vs X^2')
plt.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x, label='X vs X')
axes.plot(x, x**2, label = 'X vs X^2')
plt.legend(loc = 'upper right')
```

### Colours and Styles

```python
x = np.linspace(0, 11, 10)
```

```python
x
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color = 'blue')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color = 'red')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color = '#650dbd')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x, color = '#e6c017')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, x, color = '#e6c017', label = 'x vs x')

axes1.plot(x, x+1, color = '#650dbd', label = 'x vs x+1')
plt.legend();
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, x, color = '#e6c017', ls = '--', lw = 6)

axes1.plot(x, x+1, color = '#650dbd', lw = 6)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, x, color = '#e6c017', ls = '-.', lw = 2)

axes1.plot(x, x+1, color = '#650dbd', lw = 6)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, x, color = '#e6c017', ls = '-.', lw = 2)

axes1.plot(x, x+1, color = '#650dbd', lw = 6, ls = ':')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, x, color = '#e6c017', ls = '-.', lw = 2)

axes1.plot(x, x+1, color = '#650dbd', lw = 1, marker='o')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, x, color = '#e6c017', ls = '-.', lw = 2, marker='+', ms=10)

axes1.plot(x, x+1, color = '#650dbd', lw = 1, marker='o', ls = '--')
```
