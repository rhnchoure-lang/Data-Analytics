## Conditional Filtering

- Typically in data analysis our datasets are large enough that we don't filter based on position, but instead based on a condition.
- Conditional filtering allows us to select rows based on a condition on a column.
- This leads to a discussion on organizing our data.

- Conditional Filtering:
    - Filter by single condition
    - Filter by multiple condition
    - Check against multiple possible values

```python
df = pd.read_csv('Files/tips.csv')
```

```python
df.head()
```

```python
df[df['total_bill'] >= 40]
```

```python
df[(df['total_bill'] >= 40) & (df['sex'] == 'Male')]
```

```python
df[(df['total_bill'] >= 40) & (df['sex'] == 'Female')]
```

```python
df[(df['total_bill'] >= 30) | (df['sex'] == 'Male')]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat') | (df['day'] == 'Fri')]
```

```python
df[df['day'].isin(['Sun', 'Sat', 'Fri'])]
```

### Missing Data

Options for missing data
- Keep it
- Remove it
- Replace it

### Keeping the missing data
Pros
- Easiest to do
- Does not manipulate or change the true data

Cons
- Many methods do not support NaN
- Often there are reasonable guesses 

### Dropping or Removing the missing data
Pros
- Easy to do
- Can be based on rules

Cons
- Potential to lose a lot of data or useful information
- Limits trained models for future data

### Filling in the missing data
Pros
- Potential to save a lot of data for use in training a model

Cons
- Hardest to do and somewhat arbitrary
- Potential to lead to false conclusions

### Group By Operations

A groupby() operation allows us to examine data on a per category basis. 

```python
pd.set_option('Display.max_rows', None)
```

```python
df = pd.read_csv('Files/mpg.csv')
```

```python
df.head()
```

```python
df.groupby(by = ['model_year'])[['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'origin']].mean()
```

```python
df.groupby('origin')['name'].count()
```

Aggregation Functions which can be used on group by 

- mean(): Compute mean of groups
- sum(): Compute sum of group values
- size(): Compute group sizes
- count(): Compute count of groups
- std(): Standard deviation of groups
- var(): Compute variance of groups
- sem(): Standard error of the mean of groups
- describe(): Generates descriptive statistics
- first(): Compute first of group values
- last(): Compute last of group values
- nth: Take nth value or a subset if n is a list
- min(): Compute min of group values
- max(): Compute max of group values

### Combining DataFrames - Inner Merge

- The .merge() method takes in a key argument labelled how
- There are 3 main ways of merging tables together using the how parameter
    - Inner
    - Outer
    - Left or Right

```python
registrations = pd.DataFrame({'reg_id': [1, 2, 3, 4], 'name': ['Andrew', 'Bobo', 'Claire', 'David']})
logins = pd.DataFrame({'log_id': [1, 2, 3, 4], 'name': ['Xavier', 'Andrew', 'Yolanda', 'Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations, logins, how='inner', on='name')
```

### Combining DataFrames - Left and Right Merge

```python
pd.merge(left=registrations, right=logins, how='left', on='name')
```

```python
pd.merge(left=registrations, right=logins, how='right', on='name')
```

### Combining DataFrames - Outer Merge

```python
pd.merge(left=registrations, right=logins, how='outer', on='name')
```

```python
pd.merge(right=registrations, left=logins, how='outer', on='name')
```

```python
pd.merge(registrations, logins, how='outer', on='name')
```

```python
registrations = registrations.set_index('name')
```

```python
registrations
```

```python
pd.merge(registrations, logins, how='outer', left_index=True, right_on='name')
```

```python
registrations = registrations.reset_index()
```

```python
registrations.columns = ['reg_name', 'reg_id']
registrations
```

```python
pd.merge(registrations, logins, how='outer', left_on='reg_name', right_on='name')
```

```python
pd.merge(registrations, logins, how='inner', left_on='reg_name', right_on='name')
```

### Pivot Tables
Pivot tables allows you to reorganize data, refactoring cells, based on columns and a new index.

```python
df = pd.read_csv('Files/Sales_Funnel_CRM.csv')
```

```python
df.head()
```

```python
licenses = df[['Company', 'Product', 'Licenses']]
```

```python
licenses
```

```python
pd.pivot(data=licenses, index='Company', columns='Product', values='Licenses')
```

```python
pd.pivot_table(df, index='Company', aggfunc='sum')
```

```python
pd.pivot_table(df, index='Company', aggfunc='sum')[['Licenses', 'Sale Price']]
```

```python
df.groupby(['Company'])[['Licenses', 'Sale Price']].sum()
```

```python
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], aggfunc='sum', columns='Product', fill_value=0)
```

```python
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], aggfunc=[np.sum, np.mean], columns='Product', fill_value=0)
```

```python
pd.pivot_table(df, index=['Account Manager', 'Contact', 'Product'], values=['Sale Price'], aggfunc=[np.sum, np.mean], fill_value=0)
```
